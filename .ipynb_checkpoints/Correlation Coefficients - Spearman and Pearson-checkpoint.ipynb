Correlation coefficients are used in statistics to measure how strong a relationship is between two variables. 

## How to calculate Pearson Corellation?

Watch this video:
https://www.youtube.com/watch?v=fbY45mz42so


Calculate the mean of X, and mean of Y. Split the 2d space by the (mean(X),mean(Y)) dot (like in the pic). 
Every dot that sits in:
    Top right corner +
    Top left corner -
    Bottom right -
    Bottom left +

We will do the calculation on the std plot for each dot to normalize the values.
So the new plot represent the number of std shifts for each point on the X and Y axis.

For eample, a point (10,20) where the mean is (5,10) and std is (2,10) will transform to (2.5,1) in the new plot. <br>
Zx = (x-mean(x))\std(x) = 2.5 <br>
Zy = (y-mean(y))\std(y) = 1.

We wil multiple Zx by Zy for each point, sum all, devide by number of points-1, and you got a Pearson correlation.<br>
(if you want to understand why n-1 watch this: https://www.youtube.com/watch?v=ANsVodOu1Tg)


## Notes about Pearson correlation:

1. Can only tells us the direction and strength of <b>linear</b> relationships.
2. Dont tell us about the FORM! (linear, polinomial etc).
3. It is between -1 and 1.
4. Its the same for XvsY and YvsX.
5. Its not affected by shifting or scaling the data
6. Has no units!
7. Sensitive to outliers!

Here is some code to play with correlation plots. Change the coff variable.

```python
import numpy as np
from matplotlib.pyplot import scatter
import matplotlib.pyplot as plt


corr = 0.9         # correlation
num_samples = 200
xx = np.array([-10,10])  # X axis range from -10 to 10
yy = np.array([10,50])   # Y axis range from 10 to 50
means = [xx.mean(), yy.mean()]  
stds = [xx.std() / 3, yy.std() / 3]
covs = [[stds[0]**2          , stds[0]*stds[1]*corr], 
        [stds[0]*stds[1]*corr,           stds[1]**2]] 

m = np.random.multivariate_normal(means, covs, num_samples).T
scatter(m[0], m[1])
plt.show()
```
